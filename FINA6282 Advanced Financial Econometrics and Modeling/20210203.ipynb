$\eta_{t|t-1} \sim \mathrm{i.i.d.}\;\mathrm{N}(0, f_{t|t-1})$

$g(\eta_{t|t-1})$ is the p.d.f. of $\eta_{t|t-1}$.

$g(\eta_{t|t-1}) = \frac{1}{\sqrt{2\pi f_{t|t-1}}} \cdot \mathrm{e}^{-\frac{\big(\eta_{t|t-1} - 0\big)^2}{2f_{t|t-1}}}.$

Likelihood function is:
$\mathcal{L} = g(\eta_{1|0}) \cdot g(\eta_{2|1}) \cdot \cdots \cdot g(\eta_{T|T-1}).$

Most often, it is easier to maximize $\ln(\mathcal{L})$, which will yield the same solution (because log is a monotonic function).

$\ln(A\cdot B)  = \ln A + \ln B.$

So, $\ln(\mathcal{L}) = \ln g(\eta_{1|0}) + \ln g(\eta_{2|1}) + \cdots$ 

$g(\eta_{t|t-1}) = \frac{1}{\sqrt{2\pi f_{t|t-1}}} \cdot \mathrm{e}^{-\frac{\big(\eta_{t|t-1} - 0\big)^2}{2f_{t|t-1}}}.$

$\ln g(\eta_{t|t-1}) = -\frac{1}{2}\ln(2\pi f_{t|t-1}) -\frac{\eta_{t|t-1}^2}{2f_{t|t-1}} $

MLE maximizes $\sum_{t=1}^T \ln g(\eta_{t|t-1})$,

OLS minimizes $\sum_{t=1}^T \eta^2_{t|t-1}$.


$\ln \mathcal{L} = -\frac{1}{2} \sum_{t=1}^T \ln(2\pi |\mathbf{f}_{t|t-1}|) -\frac{1}{2} \sum_{t=1}^T\pmb{\eta}_{t|t-1}'\mathbf{f}^{-1}_{t|t-1}\pmb{\eta}_{t|t-1}.$

For our time-varying model, the formula will be:

$\ln \mathcal{L} = -\frac{1}{2}\sum_{t=1}^T \ln(2\pi f_{t|t-1}) -\frac{1}{2} \sum_{t=1}^T \eta_{t|t-1}^2 f^{-1}_{t|t-1} $.

Random walk: $y_t = y_{t-1} + e_t$.

Random walk with a drift: $y_t = \delta + y_{t-1} + e_t.$
