```python
class a_triangle:
    a = None
    b = None
    c = None
    perimeter = None
    area = None
    def __init__(self,side1,side2,side3):
        self.a = side1
        self.b = side2
        self.c = side3
    def getperim(self):
        self.perimeter = self.a + self.b + self.c
    def getarea(self):
        p = self.perimeter/2
        self.area = sqrt(p*(p-self.a)*(p-self.b)*(p-self.c))
```

<strong><u>Note</u></strong>:<br><br>
$P = a + b +c$ <br>
$p = P/2$<br>
$area = \sqrt{p\cdot(p-a)\cdot(p-b)\cdot(p-c)}$

```python
class a_figure:
    def __init__(self,params):
        self.params = params
    def getperim(self):
        self.perimeter = sum(self.params)
        return self.perimeter

class a_rectangle(a_figure):
    def getarea(self):
        self.area = self.params[0] * self.params[1]
        return self.area

class a_triangle(a_figure):
    def getarea(self):
        p = self.getperim()/2
        self.area = m.sqrt(p*(p-self.params[0])*(p-self.params[1])*(p-self.params[2]))
        return self.area
```

```python
class a_figure:
    def __init__(self,params):
        self.params = params
    def getperim(self):
        self.perimeter = sum(self.params)
        return self.perimeter

class a_rectangle(a_figure):
    def getarea(self):
        self.area = self.params[0] * self.params[1]
        return self.area

class a_triangle(a_figure):
    def getarea(self):
        p = self.getperim()/2
        self.area = m.sqrt(p*(p-self.params[0])*(p-self.params[1])*(p-self.params[2]))
        return self.area
    
class a_circle(a_figure):
    def getperim(self):
        self.perimeter = 2 * m.pi * self.params[0]
        return self.perimeter
    def getarea(self):
        self.area = m.pi * self.params[0]**2
        return self.area
```

```python
c = a_circle([5])
```
